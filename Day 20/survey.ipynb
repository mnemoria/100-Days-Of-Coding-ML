```python
country_respondents = df['Country'].value_counts()
country_respondents
```

```python
country_uses_python = country_grp['LanguageWorkedWith'].apply(lambda x: x.str.contains('Python').sum())
country_uses_python
```

```python
python_df = pd.concat([country_respondents, country_uses_python], axis='columns', sort=False)
python_df
```

```python
python_df.rename(columns={'Country':'TotalRespondents', 'LanguageWorkedWith':'WorkedWithPython'}, inplace=True)
python_df
```

```python
python_df['KnowsPythonRate'] = (python_df['WorkedWithPython'] / python_df['TotalRespondents']) * 100
python_df
```

```python
python_df.sort_values(by='KnowsPythonRate', ascending=False, inplace=True)
python_df.head()
```
